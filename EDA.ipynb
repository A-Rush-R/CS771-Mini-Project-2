```python
import torch
import matplotlib.pyplot as plt
import random
import os
```

```python
train_dir = os.path.join('dataset', 'part_one_dataset', 'train_data')
train_path = os.path.join(train_dir, '1_train_data.tar.pth')
```

```python
t = torch.load(train_path, weights_only = False)
print('the keys are', t.keys()) # it will print data and targets
data, targets = t['data'], t['targets'] # both numpy.ndarray
```

```python

print('data shape is', data.shape);
print('targets shape is', targets.shape);
```

```python
img, target = random.choice(list(zip(data, targets)))
plt.figure(figsize=(5,5))
plt.imshow(img)
plt.title('target is {}'.format(target))
plt.show()
```

```python
print(f'unique labels are {set(targets)}')
```
